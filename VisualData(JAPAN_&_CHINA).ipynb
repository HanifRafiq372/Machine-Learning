```python
#!pip install warnings
!pip install plotly.express
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
import warnings

# To display visuals within the notebook
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
# Load the dataset from the provided file path
file_path = '/content/Big_Japan_vs_China_Technology.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head(10)
```

```python
# Get an overview of the dataset
df.info()

# Summary statistics for numerical columns
df.describe()
```

```python
# Check for any missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]  # Display columns with missing values
```

```python


# Fill missing values with median or mean based on column types
df['5G Network Coverage (%)'] = df['5G Network Coverage (%)'].fillna(df['5G Network Coverage (%)'].mean())
df['Number of Startups'] = df['Number of Startups'].fillna(df['Number of Startups'].median())

# Verify missing values have been handled
df.isnull().sum().sum()  # Should return 0 if no missing values remain
```

```python


# Select only numerical columns for correlation calculation
numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Indicators')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Number of Patents Filed (Annual)', hue='Country', style='Country', markers=True)
plt.title('Number of Patents Filed Over Time (China vs Japan)')
plt.xlabel('Year')
plt.ylabel('Number of Patents Filed')
plt.grid(True)
plt.show()
```

```python
# Group by Tech Sector and Country
sector_share = df.groupby(['Tech Sector', 'Country'])['Market Share (%)'].mean().unstack()

# Plot stacked bar chart
sector_share.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Paired')
plt.title('Market Share Distribution by Tech Sector (China vs Japan)')
plt.ylabel('Average Market Share (%)')
plt.legend(title='Country')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='R&D Investment (in USD)', y='Number of Patents Filed (Annual)', hue='Country', size='Market Share (%)', palette='muted', sizes=(40, 400))
plt.title('R&D Investment vs Patents Filed (China vs Japan)')
plt.xlabel('R&D Investment (USD Billions)')
plt.ylabel('Number of Patents Filed (Annual)')
plt.show()
```

```python
china_top_exports = df[df['Country'] == 'China']['Top Tech Products Exported'].value_counts().head(10)

plt.figure(figsize=(10, 6))
china_top_exports.plot(kind='bar', color='teal')
plt.title('Top Tech Products Exported by China')
plt.xlabel('Tech Products')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
japan_top_exports = df[df['Country'] == 'Japan']['Top Tech Products Exported'].value_counts().head(10)

plt.figure(figsize=(10, 6))
japan_top_exports.plot(kind='bar', color='orange')
plt.title('Top Tech Products Exported by Japan')
plt.xlabel('Tech Products')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Venture Capital Funding (in USD)', hue='Country', marker="o", palette='Set1')
plt.title('Venture Capital Funding Over Time (China vs Japan)')
plt.xlabel('Year')
plt.ylabel('Venture Capital Funding (USD Billions)')
plt.show()
```

```python
china_market_share = df[df['Country'] == 'China'].groupby('Tech Sector')['Market Share (%)'].mean()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(china_market_share, labels=china_market_share.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('coolwarm', len(china_market_share)))
plt.title('China: Market Share by Tech Sector')
plt.show()
```

```python
japan_market_share = df[df['Country'] == 'Japan'].groupby('Tech Sector')['Market Share (%)'].mean()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(japan_market_share, labels=japan_market_share.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3', len(japan_market_share)))
plt.title('Japan: Market Share by Tech Sector')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Internet Penetration (%)', hue='Country', kde=True, palette='Set2')
plt.title('Distribution of Internet Penetration (China vs Japan)')
plt.xlabel('Internet Penetration (%)')
plt.show()
```
